```python
#   RUTA 2

import math
def dist(a,b):
    distancia=float(math.sqrt(((float(ordenes[b][1])-(float(ordenes[a][1])))**2)+((float(ordenes[b][2])-(float(ordenes[a][2])))**2)))
    distancia="{0:.1f}".format(distancia)
    return distancia
```

```python
#   RUTA 3

tmp=[]
costo=[]


for n in range(NR):
    tmp=[]
    for i in range(NR):
        tmp.append(float(dist(n,i)))
    costo.append(tmp)

print(costo)
```

```python
#   RUTA 4
PR=1

import itertools

rutas = list(itertools.permutations(nodos))
indx = 1
ruta_ord = dict()

for x in rutas:
    if PR==1:
        print("\n Ruta ",indx)
    costo_ruta =0 
    for i in range(NR-1):
        tmp = costo[x[i]][x[i+1]]
        if PR==1:
            print(x[i], "->",x[i+1] ," ", tmp) 
        costo_ruta += tmp
    if PR==1:
        print ("Costo de la ruta : ",costo_ruta)    
    key = costo_ruta*1000 + indx
    ruta_ord[key] = x
    indx += 1
    

```
